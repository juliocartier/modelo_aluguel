```python
import pandas as pd
```

```python
df = pd.read_csv('Base_de_Dados_housing_sp_city.csv')
```

```python
df
```

```python
df_locacao = df[df['tipo_anuncio'] == 'Locação']
```

```python
df_locacao
```

```python
df_locacao.columns
# numero, cep, cidade, tipo_anuncio, periodicidade, preco_venda
```

```python
df_venda = df[df['tipo_anuncio'] == 'Venda']
```

```python
df_venda.columns
# numero, cep, cidade, tipo_anuncio, periodicidade, preco_aluguel
```

```python
df_venda
```

```python
total_locacao = df_locacao.isnull().sum().sort_values(ascending=False)
percent_1_locacao = df_locacao.isnull().sum() / df_locacao.isnull().count()*100
percent_2_locacao = (round(percent_1_locacao, 2)).sort_values(ascending=False)
missing_data_locacao = pd.concat([total_locacao, percent_2_locacao], axis=1, keys=['Total de Nulos', '%'])
missing_data_locacao = missing_data_locacao.sort_values(by=['Total de Nulos'], ascending=False)
missing_data_locacao
```

```python
total_venda = df_venda.isnull().sum().sort_values(ascending=False)
percent_1_venda = df_venda.isnull().sum() / df_venda.isnull().count()*100
percent_2_venda = (round(percent_1_venda, 2)).sort_values(ascending=False)
missing_data_venda = pd.concat([total_venda, percent_2_venda], axis=1, keys=['Total de Nulos', '%'])
missing_data_venda = missing_data_venda.sort_values(by=['Total de Nulos'], ascending=False)
missing_data_venda
```

```python
## REMOVER AS COLUNAS PARA LOCACAO DOS IMOVEIS
df_locacao = df_locacao.drop(columns=['numero', 'cep', 'cidade', 'tipo_anuncio', 'periodicidade', 'preco_venda'])
```

```python
## REMOVER AS COLUNAS PARA VENDAS DOS IMOVEIS
df_venda = df_venda.drop(columns=['numero', 'cep', 'cidade', 'tipo_anuncio', 'periodicidade', 'preco_aluguel'])
```

```python
df_locacao
```

```python
df_venda
```

```python
df_locacao = df_locacao.rename({'preco_aluguel': 'preco'}, axis=1)
df_locacao
```

```python
df_venda = df_venda.rename({'preco_venda': 'preco'}, axis=1)
df_venda
```

```python
df_locacao = df_locacao.drop(columns=['taxa_condominio', 'iptu_ano'])
```

```python
df_venda['taxa_condominio'] = df_venda['taxa_condominio'].fillna(0)
df_venda['iptu_ano'] = df_venda['iptu_ano'].fillna(0)
#df = df.fillna(0)
```

```python
df_venda
```

```python
df_venda['preco_total'] = df_venda['preco'] + df_venda['taxa_condominio'] + df_venda['iptu_ano']
df_venda = df_venda.drop(columns=['preco', 'taxa_condominio', 'iptu_ano'])
df_venda
```
